## Part 1.1 Ceasar Cipher

```python
"""
Task:
- Implement a Caesar cipher.
- Fill in the encrypt() and decrypt() functions.
- Test with a plaintext of your choice.
Author: [Kelvin]
Course: Graduate Cryptography Lab
===================================================
"""

def caesar_encrypt(text, shift):
    text = text.lower()
    result = ""
    shift %= 26
    for ch in text:
        if ch.isalpha():
            result += chr((ord(ch) - 97 + shift) % 26 + 97)
        else:
            result += ch
    return result

def caesar_decrypt(text, shift):
    return caesar_encrypt(text, -shift)


# Example
plaintext = "The Quick Brown Fox Jumps Over The Lazy Dog"
shift = 3

cipher = caesar_encrypt(plaintext, shift)
print("Encrypted:", cipher)

plain = caesar_decrypt(cipher, shift)
print("Decrypted:", plain)
```

```python
"""
Task:
- Implement a simple monoalphabetic substitution cipher.
- Fill in the encrypt() and decrypt() functions.
- Test with a plaintext of your choice.
Author: [Michael]
Course: Graduate Cryptography Lab
===================================================
"""
# Generate a random substitution key
def generate_mono_key():
  letters = list(string.ascii_uppercase)
  shuffled = letters[:]
  random.shuffle(shuffled)
  key_map = dict(zip(letters, shuffled))
  return key_map



def encrypt_mono(plaintext, key_map):
  ciphertext = ""
  for ch in plaintext.upper():
    if ch in key_map:
      ciphertext += key_map[ch]
    else:
      ciphertext += ch # This captures non alphabetic charcters as themselves
  return ciphertext



def decrypt_mono(ciphertext, key_map):
  # TODO: Invert the key_map and decrypt
  inv_key = {}
  # <----- you are to fill in here ----->
  inv_key = {value: key for key, value in key_map.items()}
  plaintext = ""
  for ch in ciphertext.upper():
    if ch in inv_key:
      plaintext += inv_key[ch]
    else:
      plaintext += ch # This captures non alphabetic charcters as themselves
  return plaintext
```
